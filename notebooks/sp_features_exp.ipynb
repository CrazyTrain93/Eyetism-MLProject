```python
total_fix_counts_TD = df_TD['sp_fix_count']
total_fix_counts_ASD = df_ASD['sp_fix_count']

# Combine the variables into a single DataFrame
combined_counts_df = pd.concat([total_fix_counts_TD, total_fix_counts_ASD], axis=1)
combined_counts_df.columns = ['Total fixation counts Typical Developed', 'Total fixation counts ASD']

# Plot box plots to compare the two numeric variables
sns.boxplot(data=combined_counts_df)

# Add labels to the plot
plt.ylabel('Values')

# Show the plot
plt.show()


#combined_fix_counts = pd.DataFrame({'Total fixation counts Typical Developed': total_fix_counts_TD,
#                                    'Total fixation counts ASD': total_fix_counts_ASD})

#sns.boxplot(data=combined_fix_counts, x='Total fixation counts Typical Developed', y='Total fixation counts ASD')


#correlation_coeff = total_fix_counts_TD.corr(total_fix_counts_ASD)

#correlation_coeff
```

```python
total_fix_duration_TD = df_TD['sp_fix_duration_ms_total']
total_fix_duration_ASD = df_ASD['sp_fix_duration_ms_total']

# Combine the variables into a single DataFrame
combined_durations_df = pd.concat([total_fix_duration_TD, total_fix_duration_ASD], axis=1)
combined_durations_df.columns = ['Total fixation duration Typical Developed', 'Total fixation duration ASD']

# Plot box plots to compare the two numeric variables
sns.boxplot(data=combined_durations_df)

# Add labels to the plot
plt.ylabel('Values')

# Show the plot
plt.show()
```
